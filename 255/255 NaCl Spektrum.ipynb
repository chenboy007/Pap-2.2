```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

W,I=np.loadtxt("/Users/Jochen/2a).txt",unpack=True)
fehler_I=np.sqrt(I)

plt.errorbar(W,I,fehler_I, linestyle='None')
plt.title('Grafik 5: NaCl-Spektrum')
plt.xlabel('Winkel[°]')
plt.ylabel('Intensität[1/s]')

from numpy import exp, pi, sqrt
def gauss1(x, A, mu, sig,U):
    return U+(A/(sig*sqrt(pi/2)))*exp(-2*((x-mu)/sig)**2)
def gauss2(x, A, mu, sig,U):
    return U+(A/(sig*sqrt(pi/2)))*exp(-2*((x-mu)/sig)**2)
def gauss3(x, A, mu, sig,U):
    return U+(A/(sig*sqrt(pi/2)))*exp(-2*((x-mu)/sig)**2)
def gauss4(x, A, mu, sig,U):
    return U+(A/(sig*sqrt(pi/2)))*exp(-2*((x-mu)/sig)**2)

popt1,pcov1=curve_fit(gauss1,W[17:20],I[17:20],sigma=fehler_I[17:20],p0=[1000,6.4,0.2,650])
x1=np.linspace(6.1,7,100)
popt2,pcov2=curve_fit(gauss2,W[21:24],I[21:24],sigma=fehler_I[21:24],p0=[841,7.2,0.2,450])
x2=np.linspace(6.83,7.7,100)
popt3,pcov3=curve_fit(gauss3,W[47:53],I[47:53],sigma=fehler_I[47:53],p0=[71,12,0.2,150])
x3=np.linspace(12,13.5,100)
popt4,pcov4=curve_fit(gauss4,W[56:61],I[56:61],sigma=fehler_I[56:61],p0=[224,167,0.3,117])
x4=np.linspace(14,16,100)

plt.plot(x1,gauss1(x1,*popt1))
plt.plot(x2,gauss2(x2,*popt2))
plt.plot(x3,gauss3(x3,*popt3))
plt.plot(x4,gauss4(x4,*popt4))
```
